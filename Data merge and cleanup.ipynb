```python
# Dependencies and Setup
import pandas as pd
import datetime

```

```python
# File to Load (Remember to Change These)
file_to_load = "store.csv"

# Read Purchasing File and store into Pandas data frame
store_data = pd.read_csv(file_to_load)
store_data
```

```python
# File to Load (Remember to Change These)
file_to_load = "train.csv"

# Read Purchasing File and store into Pandas data frame
train_data = pd.read_csv(file_to_load)
train_data
```

```python
#look for missing values
train_data.count()
```

```python
# add columns from store df to train df, joining on column Store
train_merged = pd.merge(train_data, 
                  store_data[['Store', 'CompetitionDistance', 'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', 'Promo2']],    
                  on='Store')
train_merged.CompetitionDistance.fillna(train_merged.CompetitionDistance.mode()[0], inplace=True)
train_merged.CompetitionOpenSinceMonth.fillna(train_merged.CompetitionOpenSinceMonth.mode()[0], inplace=True)
train_merged.CompetitionOpenSinceYear.fillna(train_merged.CompetitionOpenSinceYear.mode()[0], inplace=True)
train_merged.CompetitionOpenSinceMonth = train_merged.CompetitionOpenSinceMonth.astype(int)
train_merged.CompetitionOpenSinceYear = train_merged.CompetitionOpenSinceYear.astype(int)

train_merged.head()
```

```python
# Extract year and month from Date
train_merged['month'] = pd.DatetimeIndex(train_merged['Date']).month
train_merged['year'] = pd.DatetimeIndex(train_merged['Date']).year
train_merged.head()

```

```python
# File to Load (Remember to Change These)
file_to_load = "test.csv"

# Read Purchasing File and store into Pandas data frame
test_data = pd.read_csv(file_to_load)
test_data.head()
```

```python
# drop id column
test_data = test_data.drop(columns=["Id"])
test_data.head()
```

```python
#look for missing values
test_data.count()
```

```python
#drop null values
test_data = test_data.dropna(how='any')
```

```python
#confirm records with missing values are removed
test_data.count()
```

```python
# add columns from store df to test df, joining on column Store
test_merged = pd.merge(test_data, 
                  store_data[['Store', 'CompetitionDistance', 'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', 'Promo2']],    
                  on='Store')
test_merged.CompetitionDistance.fillna(test_merged.CompetitionDistance.mode()[0], inplace=True)
test_merged.CompetitionOpenSinceMonth.fillna(test_merged.CompetitionOpenSinceMonth.mode()[0], inplace=True)
test_merged.CompetitionOpenSinceYear.fillna(test_merged.CompetitionOpenSinceYear.mode()[0], inplace=True)
test_merged.CompetitionOpenSinceMonth = test_merged.CompetitionOpenSinceMonth.astype(int)
test_merged.CompetitionOpenSinceYear = test_merged.CompetitionOpenSinceYear.astype(int)

test_merged
```

```python
# Extract year and month from Date
test_merged['month'] = pd.DatetimeIndex(test_merged['Date']).month
test_merged['year'] = pd.DatetimeIndex(test_merged['Date']).year
test_merged.head()
```

```python
# chcek nan values
test_merged.isnull().sum().sum()
```

```python
#finalize dfs and write to csv

train_merged = train_merged[["Store", "DayOfWeek", "Date", "month", "year", 
                             "Sales", "Customers", "Open",
                             "Promo", "StateHoliday", "SchoolHoliday", "CompetitionDistance", 
                             "CompetitionOpenSinceMonth", "CompetitionOpenSinceYear", "Promo2"]]
train_merged.head()

```
